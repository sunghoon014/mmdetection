```python
import os.path as osp

import matplotlib.pyplot as plt
import mmcv
import numpy as np

from mmcv import Config
from mmdet.datasets import build_sidewalk_dataset
```

```python
cfg = Config.fromfile('configs/NIA_sidewalk_dataset/mask_rcnn_r50_fpn_1x.py')
train_dataset = build_sidewalk_dataset(cfg.data.train, mode='train')
```

```python
import random
rand_img_id = random.choice(train_dataset.coco.getImgIds())
img = train_dataset.show_annotations(rand_img_id, show=False, out_file=None)
```

```python
plt.figure(figsize=(15, 10))
plt.imshow(mmcv.bgr2rgb(img))
```
